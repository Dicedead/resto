```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from flep_analysis_tools import *
```

```python
unige_df = pd.read_csv("../../data/unige/UNIGE_menus.csv")
```

```python
unige_df_copy = unige_df.copy()
unige_df["ET"] = np.minimum(np.minimum(unige_df["INT"], unige_df["ET"]), unige_df["EXT"])
for i in range(len(unige_df)):
    temp = sorted([unige_df_copy["INT"][i], unige_df_copy["EXT"][i], unige_df_copy["ET"][i]])
    unige_df["INT"][i] = temp[1]
```
